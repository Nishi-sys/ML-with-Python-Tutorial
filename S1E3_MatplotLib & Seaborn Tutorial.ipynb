# I. MatplotLib Tutorial

Matplotlib is a plotting library for Pyhton. It is the numerical mathematics extension of NumPy. It provides an object-oriented API for embedding plots into applications using general purpose GUI toolkits like Tkinter, wxPython, Qt or GTK+.

Pros:

Supports custom labels and texts, Genral control over every element in figure, high quality output in many formats and very customizable in general

## Creating subplots

```python
plt.subplot(2,2,1)
# 4 plots in 2 columns and 2 rows with this plot at index1
# Index: 1   2
#        3   4
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')

#If both are given for the same index, then the last one wins and overrides the previous one
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
```

```python
# Compute the x and y coordinates for points on a sine curve

x=np.arange(0,4*np.pi,2)
#Adding an interval of 0.1 smoothens the curve as more data points are plotted
#Adding 2 skews, see the difference with 0.1
y=np.sin(x)
plt.title("sine-wave funtion")
plt.plot(x,y)
plt.show()
```

```python
# Compute the x and y coordinates for points on a sine curve

x=np.arange(0,4*np.pi,0.1)
#Adding an interval of 0.1 smoothens the curve as more data points are plotted
#Adding 2 skews, see the difference with 0.1
y=np.sin(x)
plt.title("sine-wave funtion")
plt.plot(x,y)
plt.show()
```

```python
#subplot()
#Compute x and y co-ordinates for points on sine and cosine curves
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

#Set up a subplot grid that has height 2 and width1

plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sine Curve')

#Set the second subplot as active and make the second sub plot
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine Curve')

#Show the curve
plt.show()
```

## Box Plot using Matplotlib

```python
data = [np.random.normal(0,std,100) for std in range (1,4)]
#Selecting a normal distribution between some lower value to the 
#to the standard deviation value with step size as 100
#This std will loop from 1 to 4

#In the first loop std betn. 0 to 1
#In the second loop std betn. 0 to 2
#In the third loop std betn. 0 to 3
#Each with step size of 100

#rectangular boxplot
#Box plots helps you to find percentiles
plt.boxplot(data,vert=True,patch_artist=True);

#patch_artist = False will give no colour
#vert=False will give horizontal box plots

#The small circles in the figure below are the outliers
```

```python
data

#Three list of values will get created and each list will have a different array
```

## Pie Chart


```python
#Data to plot

labels ='Python','C++','Ruby','Java'
sizes=[215,130,245,210]  #Based on cumulative total
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0.2,0) #explode 1st slice

#How much the first slice will move?
```

```python
#Plot

plt.pie(sizes,explode=explode,labels=labels,colors=colors,
       autopct='%1.1f%%',shadow=True)

#auto_pct: all the percentages will be shown in floating format
#shadow=True for the shadows in slice
plt.axis('equal')
plt.show()
```

# II. Seaborn Tutorial

Distribution plots:
Seaborn is backbone of EDA as it has a lot of statistical tools

In Supervised ML, need to divide dataset in dependent and independent features.
So you need to comapre how two variable(2d)/threevariable(3d)/four variables are dependent on each other(4d).

To see change in one variable: Univariate analysis

Comparing two variables or relation between two variabes: Bivariate analysis. So these plots will be helpful

:pairplot (for more than 3features, where 4d and 5d diagram are not possible, we use pairplot)

## Correlation with Heatmap

A correlation heatmap uses colored cells, typically in a monochrmoatic scale, to show a 2D correlation matrix(table) between twp discrete dimensions or event types. Its an important featrue in Feature Selection.

Correlation or interrelation between the varaibles are found for integer and float types and not for string types. Correlation ranges from -1 to +1


```python
df.corr()

#for 0.675734, as total bill is increases, tip also increases
#Had it been -0.675734. then if total bill is decreases, then tip also decreases
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr())    #for heatmap 
                          #We need to give the matrix of correlation
```

This type of correlation is important in EDA where we do feature selection. Suppose there are two independent features and the correlation of these two independent features is same or in similar range, then you can just use one feature instead of two.

## Jointplot

A joinplot allows tostudy the relationship between 2 numeric variabels.The central chart displays thier correlation.It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D desnity plot.

### Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')

#Points will be shown as hexagonal markers
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')

#With kind='reg',it will draw a probability distribution function(PDE)
#It will also draw the regressin line
```

### Pairplot:

A 'pairplot' is also known as scatterplot, in which one variable in the same data row is matched with another variables' value, like this: Pair plots are just elaborations on this, showing all variables paired with all other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')

#To classify the plots according to a categorical variable
#In this case the categorical variable is 'sex'
```

```python
df['smoker'].value_counts()
```

### Distplot:

Dist plot helps us to check the distribution of the columns feature:

```python
sns.distplot(df['tip'])

#Generates a histogram for the tips
```

```python
sns.distplot(df['tip'],kde=False, bins=20)

#With kde=false (Kernel Density Estimation), it gives the actual values in the y-axis
#By default kde is true (which gives y asis as percentages)
```
